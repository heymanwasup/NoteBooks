```python
import ROOT as R
```

```python
# result_28paras_cbo_lost_vw_expansion_T1950
# threshold scan

f_TScan = R.TFile('/Users/cheng/workspace/TestCMake_fitter/run/threshold_scan_10pars.root')
graph = R.TGraph()
for n in range(1200,2200,50):
    func = f_TScan.Get('func_10paras_cbo_lost_T%s'%(n))
    dR = func.GetParError(3)
    graph.SetPoint(graph.GetN(),n,1./dR)

c = R.TCanvas()
l = R.TLine(1650,0.8,1650,0.9433764616445826)
graph.SetTitle('T-method threshold scan;E [MeV];1/#delta w')
graph.Draw()
l.Draw()
l.SetLineColor(R.kRed)

c.Draw()
```

```python
# result_28paras_cbo_lost_vw_expansion_T1950
# threshold scan

f_EScan = R.TFile('/Users/cheng/workspace/TestCMake_fitter/run/slices_scan_10pars.root')

graph = R.TGraphErrors()
for n in range(1000,3100,100):
    func = f_EScan.Get('func_10paras_cbo_lost_E%s_%s'%(n,n+100))
    # print ('func_10paras_cbo_lost_E%s_%s'%(n,n+100))
    # func_10paras_cbo_lost_E2900_3000;1
    kloss_v = func.GetParameter(9)
    kloss_e = func.GetParError(9)
    N = graph.GetN()
    graph.SetPoint(N,n+50,kloss_v*1e-9)
    graph.SetPointError(N,0,kloss_e*1e-9)
    # print (kloss_v,kloss_e)

c = R.TCanvas()
# l = R.TLine(linex,0.8,linex,0.9433764616445826)
graph.SetMarkerSize(0.9)
graph.SetLineStyle(1)
graph.SetLineWidth(1)
graph.SetMarkerStyle(8)
graph.SetTitle('Energy slcies scan;E [MeV];kLoss')
graph.Draw('AP')
# l.Draw()
# l.SetLineColor(R.kRed)


c.Draw()
```
