```python
import ROOT as R
```

```python
h = R.TH1D('test','',10,0,10)
r = R.TRandom3()
for n in range(1000):
    v = r.Gaus(0,1)
    h.Fill(v)
```

```python
line = R.TF1('line','[0]*sin(x)',0,10)
res = h.Fit(line,'REMS')
c = R.TCanvas()
h.Draw()
c.Draw()
```

```python
print (res.IsValid())
print (int(res))
print (res.Status())
```
